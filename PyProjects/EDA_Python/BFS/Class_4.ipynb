```python
from collections import deque
def search (name):
    search_queue = deque()
    search_queue += grafo[name]
    searched = []
    while search_queue:
        print(search_queue)
        person = search_queue.popleft()
        if not person in searched:
            if person == "Thom":
                print(person + " es vendedor")
                return True
            else:
                search_queue += grafo[person]
                searched.append(person)
    return False
```

```python
# Ejemplo llegada de Twin al Golden 
grafo_2 = {}
grafo_2["TWIN"] = ["WALK1", "WALK2"]  # es importante mantener el orden de la lista de las values, porque es una lista y y tiene indices
grafo_2["WALK1"] = ["BUS44"]
grafo_2["WALK2"] = ["BUS33", "BUS33_B"]
grafo_2["BUS33"] = ["BUS44"]
grafo_2["BUS33_B"] = ["BUS44"]
grafo_2["BUS44"] = ["GOLDEN"]
grafo_2["GOLDEN"] = []
```

```python
from collections import deque
def search_path(start, finish):
    search_queue = deque()
    search_queue.append(start)
    edgeTo = {start: None}  # Diccionario para registrar el padre de cada nodo
    searched = []

    while search_queue:
        word = search_queue.popleft()
        if word == finish:
            print(word + " es el nodo finish, se ha encontrado un path")
            # Reconstruir la ruta desde el objetivo hasta el inicio
            path = []
            while word is not None:
                path.append(word)
                word = edgeTo[word]
            path.reverse()  # Invertir para obtener la ruta desde el inicio
            print("Ruta encontrada:", " -> ".join(path))
            return True
        else:
            for neighbor in grafo_2[word]:
                if neighbor not in searched:
                    search_queue.append(neighbor)
                    edgeTo[neighbor] = word  # Registrar el nodo padre
            searched.append(word)

    print("No se encontrÃ³ un path hasta el nodo finish.")
    return False
```

```python
search_path("TWIN", "GOLDEN")
```

Este algoritmo tiene su principal aplicacion en el recorrido de laberintos, siendo cada interseccion un nodo.

El objetivo es alejarse de los nodos ya visitados o de los iniciales, si se vuelve a llegar a los nodos iniciales se retrocede para buscar otros caminos, se retrocede especialmente si no se tiene mas caminos
