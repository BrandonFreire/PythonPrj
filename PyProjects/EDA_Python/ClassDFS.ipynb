```python
def dfs(nodo, grafo, searched, componentR, parent):
    componentR.append(nodo) # ir agregando los nodos que pertenecen al componente conectado
    searched [nodo] = True # marcar al nodo como visitado

    print(searched)
    print('Vecinos de', nodo, ':', grafo[nodo]) #qué vecinos tiene el nodo

    # Recorrer los nodos adyacentes del nodo dado
    for vecino in grafo [nodo]:
        if not searched [vecino]: # verificar que el nodo adyacente no esta marcado como searched
            parent[vecino] = nodo
            dfs(vecino, grafo, searched, componentR, parent) # llamada recursiva DFS
            print('Finaliza', vecino)
            print('Vuelve a', nodo)
            print()
    print(parent)
```
